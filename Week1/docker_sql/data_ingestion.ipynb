## Data Ingestion

So we can now create the table and inject the data into it.

```python
from sqlalchemy import create_engine
import pyarrow.parquet as pq
from time import time
```

```python
engine = create_engine('postgresql://postgres:postgres@localhost:5432/ny_taxi', echo=False)
```

We will iterate over the dataset and insert the data into the table.

```python
yellow_pf = pq.ParquetFile('data/yellow_tripdata_2019-10.parquet')
green_pf = pq.ParquetFile('data/green_tripdata_2019-10.parquet')
```

```python
def process_taxi_data_batches(pf, table_name, engine, batch_size=100000):
    total_time = 0
    total_data = 0
    for i, batch in enumerate(pf.iter_batches(batch_size=100000)):
        start = time()
        batch.to_pandas().to_sql(table_name, engine, if_exists='append')
        end = time() - start
        print(f'Batch: {i}. Elapsed time: {end:.2f} sec')
        total_time += end
        total_data += len(batch)
    print(f'Total time: {total_time:.2f} sec. Total data: {total_data}')

process_taxi_data_batches(yellow_pf, 'yellow_taxi_2019', engine)
process_taxi_data_batches(green_pf, 'green_taxi_2019', engine)
```

## Zones

Now we will create the zones table and insert the data into it.

```python
zones_pf = pq.ParquetFile('data/taxi_zone_lookup.parquet')
```

```python
process_taxi_data_batches(zones_pf, 'zones', engine)
```
